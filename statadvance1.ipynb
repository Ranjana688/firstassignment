13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The Central Limit Theorem (CLT) states that the distribution of sample means will approximate a normal distribution, regardless of the original population's distribution, as long as the sample size is sufficiently large. This is critical in statistics because it allows us to make inferences about population parameters (like the mean) using sample data, even when the population distribution is unknown or non-normal.
Here's a more detailed explanation:
What it means:
Imagine you have a population with a very skewed distribution (e.g., income distribution). If you take many random samples from this population and calculate the mean of each sample, those sample means will, according to the CLT, tend to cluster around the population mean and follow a normal distribution.
Why it's important:
Many statistical methods rely on the assumption of a normal distribution. If the data you are working with doesn't appear to be normally distributed, you can still use these methods with confidence if you have a large enough sample size due to the CLT

17. What are point estimates and interval estimates in statistics?

In statistics, point estimates and interval estimates are two fundamental ways to estimate unknown population parameters using sample data
Point Estimates:
A point estimate is a single value calculated from sample data that is used to estimate an unknown population parameter.
For example, the sample mean is a point estimate for the population mean, and the sample proportion is a point estimate for the population proportion.
Point estimates are straightforward but don't provide information about the uncertainty or variability in the estimation.
Interval Estimates:
An interval estimate provides a range of values, called a confidence interval, within which the population parameter is likely to lie, with a specified level of confidence.
For example, a 95% confidence interval for the population mean would be a range of values calculated from the sample data, such that we are 95% confident that the true population mean falls within that range.


18. What is the significance of confidence intervals in statistical analysis?

Confidence intervals in statistical analysis are crucial because they provide a range of plausible values for an unknown population parameter, offering a more informative measure of uncertainty than a single point estimate

19.  What is the relationship between a Z-score and a confidence interval?
A Z-score and a confidence interval are related concepts in statistics. The Z-score, specifically the critical Z-score, is used to determine the margin of error, which in turn defines the width of the confidence interval. Essentially, the confidence interval provides a range of values likely to contain the true population parameter, and the Z-score helps define the boundaries of that range.

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
        x (int): The outcome (0 or 1)
        p (float): The probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
        float: Probability of the outcome x
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0

# Ex
p = 0.7  # Probability of success
print("P(X=0):", bernoulli_pdf(0, p))
print("P(X=1):", bernoulli_pdf(1, p))
```

9.Write a Python function to calculate Z-scores from a dataset and plot them.


```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores for the given dataset and plots them.

    Parameters:
        data (list or np.array): A numeric dataset

    Returns:
        z_scores (np.array): The calculated Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    if std_dev == 0:
        raise ValueError("Standard deviation is zero. Z-scores cannot be computed.")


    z_scores = (data - mean) / std_dev
    plt.figure(figsize=(8, 5))
    plt.plot(z_scores, 'o-', color='blue', label='Z-scores')
    plt.axhline(0, color='gray', linestyle='--')
    plt.axhline(1, color='green', linestyle='--', label='¬±1 SD')
    plt.axhline(-1, color='green', linestyle='--')
    plt.axhline(2, color='orange', linestyle='--', label='¬±2 SD')
    plt.axhline(-2, color='orange', linestyle='--')
    plt.axhline(3, color='red', linestyle='--', label='¬±3 SD')
    plt.axhline(-3, color='red', linestyle='--')
    plt.title("Z-scores of Dataset")
    plt.xlabel("Data Point Index")
    plt.ylabel("Z-score")
    plt.grid(True)
    plt.legend()
    plt.show()

    return z_scores

# Ex
data = np.random.normal(loc=50, scale=10, size=30)
z_scores = calculate_and_plot_z_scores(data)
```

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    """
    Calculates the Z-score for a data point.

    Parameters:
        x (float): The data point
        mean (float): Mean of the population
        std_dev (float): Standard deviation of the population

    Returns:
        float: Z-score
    """
    return (x - mean) / std_dev

def plot_standard_normal_with_z(z):
    """
    Plots the standard normal distribution and highlights the Z-score area.

    Parameters:
        z (float): Z-score to highlight
    """
    x = np.linspace(-4, 4, 500)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Standard Normal PDF', color='blue')
    plt.fill_between(x, 0, y, where=(x <= z), color='skyblue', alpha=0.5,
                     label=f'Area ‚â§ Z = {z:.2f}')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title('Standard Normal Distribution with Z-score')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example: given data point
x_value = 82
mean = 75
std_dev = 5

# Step 1: Calculate Z-score
z = calculate_z_score(x_value, mean, std_dev)
print(f"Z-score for x = {x_value}: {z:.2f}")

# Step 2: Calculate cumulative probability (area under curve up to Z)
cdf_value = norm.cdf(z)
print(f"Probability that Z ‚â§ {z:.2f}: {cdf_value:.4f}")

# Step 3: Visualize
plot_standard_normal_with_z(z)
```

16. Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, n, alpha=0.05, tail='two'):

    standard_error = population_std / np.sqrt(n)
    z_stat = (sample_mean - population_mean) / standard_error

    if tail == 'two':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif tail == 'right':
        p_value = 1 - norm.cdf(z_stat)
    elif tail == 'left':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("Invalid tail type. Use 'two', 'left', or 'right'.")


    if p_value < alpha:
        result = "Reject the null hypothesis (H‚ÇÄ)"
    else:
        result = "Fail to reject the null hypothesis (H‚ÇÄ)"

    return z_stat, p_value, result

# Ex
sample_mean = 104
population_mean = 100
population_std = 15
n = 50
alpha = 0.05

z_stat, p_value, decision = z_test(sample_mean, population_mean, population_std, n, alpha, tail='two')

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Decision: {decision}")
```

17. Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):

    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # sample std dev & standard error
    z = norm.ppf((1 + confidence) / 2)

    margin = z * std_err
    return (mean - margin, mean + margin)

# üî¢ Example data: sample of test scores
sample_data = [88, 92, 85, 91, 87, 89, 94, 90, 86, 93]

# Compute 95% confidence interval
ci_low, ci_high = confidence_interval(sample_data, confidence=0.95)
mean_sample = np.mean(sample_data)

# üñ®Ô∏è Print result
print(f"Sample Mean: {mean_sample:.2f}")
print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
```

18. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):

    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # sample standard error
    z = norm.ppf((1 + confidence) / 2)           # z-score for desired confidence
    margin = z * std_err

    return mean - margin, mean + margin, mean

# Step 1: Simulate normal data
np.random.seed(42)  # for reproducibility
mu = 50             # true population mean
sigma = 10          # true population std dev
n = 100             # sample size

sample_data = np.random.normal(loc=mu, scale=sigma, size=n)
ci_low, ci_high, sample_mean = confidence_interval(sample_data, confidence=0.95)
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: ({ci_low:.2f}, {ci_high:.2f})")
```

22. Write a Python program to compare the standard deviations of two datasets and visualize the difference

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=5, size=100)
data2 = np.random.normal(loc=50, scale=15, size=100)
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(['Dataset 1', 'Dataset 2'], [std1, std2], color=['steelblue', 'salmon'])
plt.title("Standard Deviation Comparison")
plt.ylabel("Standard Deviation")
plt.grid(axis='y')
plt.subplot(1, 2, 2)
plt.hist(data1, bins=20, alpha=0.6, label='Dataset 1 (std ~5)', color='steelblue', edgecolor='black')
plt.hist(data2, bins=20, alpha=0.6, label='Dataset 2 (std ~15)', color='salmon', edgecolor='black')
plt.title("Histogram of Two Datasets")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

23. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

```python
import numpy as np

# Step 1: Generate data from a normal distribution
np.random.seed(42)
mean = 100
std_dev = 15
n = 100
data = np.random.normal(loc=mean, scale=std_dev, size=n)
data_range = np.max(data) - np.min(data)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
print(f"Sample Size: {n}")
print(f"Data Range: {data_range:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

24. Implement Z-score normalization on a dataset and visualize its transformation

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a sample dataset
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=100)  # mean=50, std=10
mean = np.mean(data)
std = np.std(data)
z_normalized = (data - mean) / std
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(mean, color='red', linestyle='--', label=f"Mean = {mean:.2f}")
plt.legend()
plt.grid(True)
plt.subplot(1, 2, 2)
plt.hist(z_normalized, bins=20, color='lightgreen', edgecolor='black')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='--', label="Mean = 0")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
print(f"Original Mean: {mean:.2f}, Std Dev: {std:.2f}")
print(f"Normalized Mean: {np.mean(z_normalized):.2f}, Std Dev: {np.std(z_normalized):.2f}")
```
